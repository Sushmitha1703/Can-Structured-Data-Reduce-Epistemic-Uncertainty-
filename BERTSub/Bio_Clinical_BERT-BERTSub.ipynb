```python
from deeponto.onto import Ontology
src_onto = Ontology("symp.owl")
tgt_onto = Ontology("CSSO_ver0.6.owl")
```

```python
subs_generator = SubsFromEquivMappingGenerator(
  src_onto, tgt_onto, equiv_mappings,
  subs_generation_ratio=1, delete_used_equiv_tgt_class=True
)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
subs_df = pd.DataFrame(subs_generator.subs_from_equivs, columns=["SrcEntity", "TgtEntity", "Score"])
train, test = train_test_split(subs_df, test_size=0.7)
train.to_csv(f"/content/refs_subs/train.tsv", sep="\t", index=False)
test.to_csv(f"/content/refs_subs/test.tsv", sep="\t", index=False)
```

```python
refs_equiv = ReferenceMapping.read_table_mappings(f"/content/drive/MyDrive/bertmap/fyp-bertmap/match/raw_mappings.tsv")
refs_subs = ReferenceMapping.read_table_mappings(f"/content/refs_subs/train.tsv")
refs_subs += ReferenceMapping.read_table_mappings(f"/content/refs_subs/test.tsv")
```

```python
from deeponto.align.mapping import *
```

```python
raw_mappings = pd.read_csv('/content/refs_equiv/raw_mappings.tsv', sep='\t')
```

```python
train_data, test_data = train_test_split(raw_mappings, test_size=0.3, random_state=42)
```

```python
train_data.to_csv('/content/refs_equiv/train.tsv', sep='\t', index=False)
test_data.to_csv('/content/refs_equiv/test.tsv', sep='\t', index=False)
```

```python
for_subsumption = True
```

```python
refs_full = refs_subs if for_subsumption else refs_equiv
```

```python
if for_subsumption:
        refs_test = ReferenceMapping.read_table_mappings(f"/content/refs_subs/test.tsv")
        result_path = f"/content/refs_subs/test.cands.tsv"
else:
        refs_test = ReferenceMapping.read_table_mappings(f"/content/refs_equiv/test.tsv")
        result_path = f"/content/refs_equiv/test.cands.tsv"
```

```python
from deeponto.align.mapping import NegativeCandidateMappingGenerator
from deeponto.align.mapping import Tokenizer
```

```python
config = BERTMapPipeline.load_bertmap_config()
neg_gen = NegativeCandidateMappingGenerator(
        src_onto, tgt_onto, refs_full,
        config.annotation_property_iris,
        Tokenizer.from_pretrained(config.bert.pretrained_path),
        for_subsumption=for_subsumption
    )
```

```python
results = []
for test_map in refs_test:
  valid_tgts, stats = neg_gen.mixed_sample(test_map, idf=50, neighbour=50)
  print(f"STATS for {test_map}:\n{stats}")
  results.append((test_map.head, test_map.tail, valid_tgts))
results = pd.DataFrame(results, columns=["SrcEntity", "TgtEntity", "TgtCandidates"])
results.to_csv(result_path, sep="\t", index=False)
```
