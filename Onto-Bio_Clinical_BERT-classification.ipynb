```python
import nltk
nltk.download('omw-1.4')
def clean(text):

  cleanr = re.compile('<[^>]*>')           # remove html
  cleantext = re.sub(cleanr, ' ', text)

  cleantext = re.sub("[-]", " " , cleantext)   # remove - sign

  cleantext = re.sub("[^A-Za-z0-9 ]", " " , cleantext)  # remove evey character except alphabet
  cleantext = cleantext.lower()

  words = nltk.tokenize.word_tokenize(cleantext)
  words_new = [i for i in words if i not in stop_words]

  w = [lemmatizer.lemmatize(word) for word in words_new if len(word)>2]

  return ' '.join(w)
```

```python
def tokenize(questions):
    encoded = tokenizer(
        questions.tolist(),
        add_special_tokens=True,
        max_length=400,
        padding='max_length',
        truncation=True,
        return_attention_mask=True,
        return_token_type_ids=True,
        return_tensors='tf'
    )
    return {
        'input_ids': encoded['input_ids'],
        'attention_mask': encoded['attention_mask'],
        'token_type_ids': encoded['token_type_ids']
    }
```

```python
def predict_user_question(user_question, model, tokenizer, label_encoder):
    # Tokenize the user's input question
    encoded_question = tokenizer.encode_plus(
        user_question,
        add_special_tokens=True,
        max_length=400,
        padding='max_length',
        truncation=True,
        return_attention_mask=True,
        return_token_type_ids=True,
        return_tensors='tf'
    )

    # Make a prediction
    predicted_labels_binary = model.predict({
        'input_22': encoded_question['input_ids'],
        'input_23': encoded_question['attention_mask'],
        'input_24': encoded_question['token_type_ids']
    })

    # Convert binary predictions to original labels
    predicted_labels_original = label_encoder.inverse_transform((predicted_labels_binary > 0.5).astype(int))
    predicted_label = ['symptom'] if predicted_labels_original > 0.5 else ['treatment']
    print(f"Original Decoded Prediction: {predicted_label}")

    return predicted_labels_original[0]

user_input_question = input("Enter your question: ")
predicted_class = predict_user_question(user_input_question, model, tokenizer, label_encoder)
print(f"The predicted class for the question '{user_input_question}' is: {predicted_class}")
```
