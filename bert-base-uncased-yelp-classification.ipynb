```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
from tqdm.notebook import tqdm_notebook
import nltk
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
```

```python
from nltk import word_tokenize
nltk.download('punkt')
nltk.download('wordnet')
```

```python
import nltk
nltk.download('omw-1.4')
def clean(text):

  cleanr = re.compile('<[^>]*>')           # remove html
  cleantext = re.sub(cleanr, ' ', text)

  cleantext = re.sub("[-]", " " , cleantext)   # remove - sign

  cleantext = re.sub("[^A-Za-z0-9 ]", " " , cleantext)  # remove evey character except alphabet
  cleantext = cleantext.lower()

  words = nltk.tokenize.word_tokenize(cleantext)
  words_new = [i for i in words if i not in stop_words]

  w = [lemmatizer.lemmatize(word) for word in words_new if len(word)>2]

  return ' '.join(w)
```

```python
filtered_df['Question'] = tqdm_notebook(filtered_df['Question'].apply(lambda x: clean(str(x))))
```

```python
filtered_df['Answer'] = tqdm_notebook(filtered_df['Answer'].apply(lambda x: clean(str(x))))
```

```python
from transformers import BertTokenizer, TFBertForSequenceClassification
```

```python
model = TFBertForSequenceClassification.from_pretrained('textattack/bert-base-uncased-yelp-polarity', num_labels=2,from_pt=True)
```

```python
def tokenize(questions):
    encoded = tokenizer(
        questions.tolist(),
        add_special_tokens=True,
        max_length=400,
        padding='max_length',
        truncation=True,
        return_attention_mask=True,
        return_token_type_ids=True,
        return_tensors='tf'
    )
    return {
        'input_ids': encoded['input_ids'],
        'attention_mask': encoded['attention_mask'],
        'token_type_ids': encoded['token_type_ids']
    }
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
train_labels = label_encoder.fit_transform(labels)
```

```python
train_tokens = tokenize(filtered_df['Question'])
```

```python
input_ids = tf.keras.Input(shape=(400,), dtype=tf.int32)
attention_mask = tf.keras.Input(shape=(400,), dtype=tf.int32)
token_type_ids = tf.keras.Input(shape=(400,), dtype=tf.int32)
```

```python
train_inputs = [train_tokens['input_ids'], train_tokens['attention_mask'], train_tokens['token_type_ids']]
```

```python
from tensorflow.keras.layers import Dense
# Unfreeze more layers for fine-tuning
for layer in model.layers[:10]:  # Adjust the number of layers to unfreeze
    layer.trainable = True

# Add dropout for regularization
dropout_layer = tf.keras.layers.Dropout(0.2)  # Adjust dropout rate as needed

input_ids = tf.keras.Input(shape=(400,), dtype=tf.int32)
attention_mask = tf.keras.Input(shape=(400,), dtype=tf.int32)
token_type_ids = tf.keras.Input(shape=(400,), dtype=tf.int32)

bert_output = model(input_ids, attention_mask=attention_mask, token_type_ids=token_type_ids)[0]
bert_output = dropout_layer(bert_output)

output = Dense(1, activation='sigmoid')(bert_output)

model = tf.keras.Model(inputs=[input_ids, attention_mask, token_type_ids], outputs=output)
```

```python
model.compile(
    optimizer=tf.keras.optimizers.AdamW(learning_rate=1e-5),  # Adjust learning rate as needed
    loss='binary_crossentropy',
    metrics=['binary_accuracy']  # Add class weights if needed
)
```

```python
model.fit(
    train_inputs,
    train_labels,
    batch_size=16,  # Adjust batch size as needed
    epochs=15,  # Increase epochs
)
```
